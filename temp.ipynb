```python
import pandas as pd
import random

# List of doctors
doctors = [
    "SELF",
    "NEELIMA AGARWAL GARU, MD(Homeo).",
    "M RAMA KRISHNA GARU, MBBS, DCH.",
    "SMT. M.N. SRI DEVI GARU, MBBS, DGO.",
    "AKULA SATYANARAYANA GARU, MBBS, DCH.",
    "SMT. A GANGA BHAVANI GARU, MBBS, DGO.",
    "K. RAJENDRA PRASAD GARU, B.M.P.",
    "CH. KIRAN KUMAR GARU, B.M.P.",
    "K. CHIRANJEEVI RAJU GARU, B.M.P.",
    "S. PRASAD GARU",
    "A. SATTI BABU GARU, MD, PGDM",
    "V RAMAKRISHNA GARU, B.M.P",
    "CH. RAGHUNADHA RAO GARU, B.M.P",
    "C.V.S.S. SHARMA GARU, BAMS",
    "B. SURIBABU GARU, B.M.P",
    "ADDANKI SURYA GARU, B.M.P",
    "AJAY KUMAR GARU, P.M.P",
    "A. CHANDRASEKHAR GARU, BSC, DPH&S",
    "DURGA PRASAD GARU, B.M.P",
    "JAGADEESH GARU, P.M.P",
    "R. RAMA CHANDRA RAO GARU, P.M.P",
    "G. BHANUPRAKASH GARU, P.M.P",
    "PV. VENKATESWARA RAO GARU, P.M.P",
    "A. GOWRISH BABU GARU, B.M.P",
    "V.V.V PRASAD GARU, P.M.P",
    "DILEEP KUMAR GARU, MBBS, MEDICAL OFFICER"
]

# Age group options and sample options
age_groups = ["Y", "M", "D"]
sample_options = ["SELF", "RAM", "RAJESH", "RAJU"]
paid_options = ["PAID", "NOT PAID", "DUE"]

# Generate DataFrame with 10 rows
data = []
for i in range(1, 11):
    amount = random.randint(500, 5000)
    paid_status = random.choice(paid_options)
    due_amount = 0
    if paid_status == "NOT PAID":
        due_amount = amount
    elif paid_status == "DUE":
        due_amount = random.randint(1, amount - 1)
    else:
        due_amount = 0

    row = {
        "S.No.": str(i),
        "Date": f"2024-{random.randint(1, 28):02d}-{random.randint(1, 12):02d}",
        "Time": f"{random.randint(1, 12):02d}:{random.randint(0, 59):02d}:{random.randint(0, 59):02d} "
                f"{random.choice(['AM', 'PM'])}",
        "Patient Name": f"Patient {i}",
        "Reference By": random.choice(doctors if i > 2 else ["SELF"]),
        "Patient Age": random.randint(1, 100),
        "Age Group": random.choice(age_groups),
        "Gender": random.choice(["Male", "Female"]),
        "Amount": amount,
        "Phone No": random.randint(7000000000, 9999999999),
        "Paid": paid_status,
        "Due": due_amount,
        "Sample": random.choice(sample_options)
    }
    data.append(row)

df = pd.DataFrame(data)
df.to_excel("assets/all_files/2024_12_01.xlsx",index=False)
```

```python
import pandas as pd
import numpy as np

# Define dtype_map and options for categorical data
dtype_map = {
    "S.No.": str,
    "Date": str,
    "Time": str,
    "Patient Name": str,
    "Reference By": str,
    "Patient Age": "Int8",
    "Age Group": str,
    "Gender": str,
    "Amount": "Int32",
    "Method": str,
    "Phone No": "Int64",
    "Paid": str,
    "Due": "Int16",
    "Sample": str
}

# Define sample data for specific fields
patient_names = ["John Doe", "Jane Smith", "Alice Brown", "Bob Davis", "Emily White"]
reference_by = ["Dr. A", "Dr. B", "Dr. C", "Dr. D", "Dr. E"]
age_groups = ["Y", "M", "D"]
genders = ["Male", "Female"]
methods = ["CASH", "PHONE PAY"]
paid_statuses = ["PAID", "NOT PAID", "DUE"]
samples = ["SELF", "RAJU", "RAJESH", "RAM"]

# Generate 20 rows of data
data = {
    "S.No.": [f"{i+1:03}" for i in range(20)],
    "Date": pd.date_range(start="2024-01-01", periods=20).strftime("%Y-%m-%d").tolist(),
    "Time": [f"{hour:02}:{minute:02} {am_pm}" for hour, minute, am_pm in zip(
        np.random.randint(1, 13, 20),
        np.random.randint(0, 60, 20),
        np.random.choice(["AM", "PM"], 20)
    )],
    "Patient Name": np.random.choice(patient_names, 20),
    "Reference By": np.random.choice(reference_by, 20),
    "Patient Age": np.random.randint(1, 100, 20),
    "Age Group": np.random.choice(age_groups, 20),
    "Gender": np.random.choice(genders, 20),
    "Amount": np.random.randint(500, 5000, 20),
    "Method": np.random.choice(methods, 20),
    "Phone No": np.random.randint(7000000000, 9999999999, 20,dtype=np.int64),
    "Paid": np.random.choice(paid_statuses, 20),
    "Due": [np.random.randint(0, 5000) if paid != "PAID" else 0 for paid in np.random.choice(paid_statuses, 20)],
    "Sample": np.random.choice(samples, 20)
}

# Create DataFrame with dtype_map
df = pd.DataFrame(data).astype(dtype_map)
df.to_csv("assets/all_files/2024_12_02.csv",index=False)
```
